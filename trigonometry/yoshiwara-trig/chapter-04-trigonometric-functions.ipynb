## The Special Angles (30°, 45°, 60°)

These angles are "special" because we can express their trig ratios exactly using radicals.

### The 30-60-90 Triangle (from equilateral triangle)

Start with an equilateral triangle (all sides equal, all angles 60°). Cut it in half:

- Original side = 2
- Half the base = 1
- Height (by Pythagorean): √(2² - 1²) = √3

**Ratio: 1 : √3 : 2** (short leg : long leg : hypotenuse)

This gives us:
- sin(30°) = 1/2, cos(30°) = √3/2, tan(30°) = 1/√3
- sin(60°) = √3/2, cos(60°) = 1/2, tan(60°) = √3

### The 45-45-90 Triangle (from isosceles right triangle)

Here's where I connect this to my perpendicular intuition from Kiselev's geometry.

---

## My Path Intuition for 45°

Remember from geometry: the perpendicular from M to N through midpoint C is the **straight path** — no turns, shortest distance.

Now let's think about it differently:

**Setup:**
- Start at point M
- Go to point D at a 45° angle (length = 1)
- Turn 90° and go to point N (length = 1, by isosceles property)
- The angle at D is 90° (two 45° angles meeting)

```
    M
     \
      \ 1 (45° angle)
       \
        D-------- (90° turn here)
         \
          \ 1
           \
            N
```

Wait, let me draw this better as an isosceles right triangle:

```
        M
        |\  
        | \  
      1 |  \ hypotenuse = ?
        |   \
        |____\ N
           1
```

**The bent path (along the legs):** M → corner → N = 1 + 1 = 2

**The straight path (hypotenuse):** M → N directly = must be LESS than 2 (triangle inequality!)

Using Pythagorean theorem:
$$\text{hypotenuse} = \sqrt{1^2 + 1^2} = \sqrt{2}$$

And indeed: √2 ≈ 1.414 < 2 ✓

The straight path saves you about 0.586 units compared to going along the legs!

### Deriving the 45° Trig Ratios

With legs = 1 and hypotenuse = √2:

$$\sin(45°) = \frac{\text{opposite}}{\text{hypotenuse}} = \frac{1}{\sqrt{2}} = \frac{\sqrt{2}}{2}$$

$$\cos(45°) = \frac{\text{adjacent}}{\text{hypotenuse}} = \frac{1}{\sqrt{2}} = \frac{\sqrt{2}}{2}$$

$$\tan(45°) = \frac{\text{opposite}}{\text{adjacent}} = \frac{1}{1} = 1$$

**Key insight:** sin(45°) = cos(45°) because the triangle is isosceles — both legs are equal!

**Ratio: 1 : 1 : √2** (leg : leg : hypotenuse)

### Why the Path Analogy Matters

The perpendicular/path intuition connects to the **triangle inequality**:

> The sum of any two sides of a triangle must be greater than the third side.

Or equivalently: **the straight line is the shortest path between two points**.

So when I have a right triangle with legs 1 and 1:
- The "bent path" along the legs = 1 + 1 = 2
- The "straight path" (hypotenuse) = √2 ≈ 1.414

This confirms √2 < 2, which is consistent with my perpendicular intuition!

The geometry proof (folding, perpendiculars) and the algebra (Pythagorean theorem) arrive at the same truth from different angles.

````python
import numpy as np
import matplotlib.pyplot as plt

# Visualize the 45-45-90 triangle with path comparison
fig, ax = plt.subplots(1, 1, figsize=(8, 8))

# The triangle vertices
M = np.array([0, 1])
corner = np.array([0, 0])
N = np.array([1, 0])

# Draw the triangle
triangle = plt.Polygon([M, corner, N], fill=False, edgecolor='blue', linewidth=2)
ax.add_patch(triangle)

# Draw the hypotenuse (straight path) in red
ax.plot([M[0], N[0]], [M[1], N[1]], 'r-', linewidth=2, label=f'Straight path: √2 ≈ {np.sqrt(2):.3f}')

# Draw the bent path in green (dashed)
ax.plot([M[0], corner[0], N[0]], [M[1], corner[1], N[1]], 'g--', linewidth=2, label='Bent path: 1 + 1 = 2')

# Mark the points
ax.plot(*M, 'ko', markersize=10)
ax.plot(*corner, 'ko', markersize=10)
ax.plot(*N, 'ko', markersize=10)
ax.annotate('M', M + [0.05, 0.05], fontsize=14)
ax.annotate('Corner (90°)', corner + [-0.15, -0.1], fontsize=12)
ax.annotate('N', N + [0.05, -0.05], fontsize=14)

# Mark the angles
ax.annotate('45°', [0.15, 0.85], fontsize=12, color='purple')
ax.annotate('45°', [0.85, 0.15], fontsize=12, color='purple')

# Add leg lengths
ax.annotate('1', [-0.1, 0.5], fontsize=14, color='blue')
ax.annotate('1', [0.5, -0.1], fontsize=14, color='blue')

ax.set_xlim(-0.3, 1.3)
ax.set_ylim(-0.3, 1.3)
ax.set_aspect('equal')
ax.legend(loc='upper right')
ax.set_title('45-45-90 Triangle: Straight Path vs Bent Path')
ax.grid(True, alpha=0.3)
plt.show()
````

## Reference Angles in All Quadrants

The special angles appear in all four quadrants. The **reference angle** is the acute angle to the x-axis.

### Example: sin, cos, tan of 210°

1. **Find the quadrant:** 210° is in Quadrant III (between 180° and 270°)
2. **Find reference angle:** θₑ = 210° - 180° = 30°
3. **Determine signs:** In Q3, both x and y are negative → sin(-), cos(-), tan(+)

$$\sin(210°) = -\sin(30°) = -\frac{1}{2}$$
$$\cos(210°) = -\cos(30°) = -\frac{\sqrt{3}}{2}$$
$$\tan(210°) = \tan(30°) = \frac{1}{\sqrt{3}}$$ (positive because -/- = +)

### Example: sin, cos, tan of 300°

1. **Quadrant:** IV (between 270° and 360°)
2. **Reference angle:** θₑ = 360° - 300° = 60°
3. **Signs:** In Q4, x is positive, y is negative → sin(-), cos(+), tan(-)

$$\sin(300°) = -\sin(60°) = -\frac{\sqrt{3}}{2}$$
$$\cos(300°) = \cos(60°) = \frac{1}{2}$$
$$\tan(300°) = -\tan(60°) = -\sqrt{3}$$
