# Section 2.4: Epsilon-Delta From First Principles

## Intuitive Notes for the Two Starter Laws and Two Worked Examples

These notes are written in a direct, informal style inspired by chalkboard explanations: meaning first, symbols second.

The point is not to memorize algebra. The point is to understand what a limit claim is promising.

---

## The Two Starter Laws

### Law 8: Constant Law

$$\lim_{x \to a} c = c$$

Intuition: if the output is always the same constant $c$, then nothing is changing as $x$ moves. There is no drift to control. The output is already exactly at $c$.

First-principles view: if we want $|c-c| < \epsilon$, that is just $0 < \epsilon$, which is true for every positive $\epsilon$. So any positive $\delta$ works.

### Law 9: Identity Law

$$\lim_{x \to a} x = a$$

Intuition: here the output is literally the input. If $x$ is close to $a$, then the output is close to $a$ for the same reason: they are the same number.

First-principles view: if we want $|x-a| < \epsilon$, we can simply choose $\delta = \epsilon$. Then $|x-a| < \delta$ automatically gives $|x-a| < \epsilon$.

These two laws matter because every polynomial and rational-function limit is built from constants, $x$, and algebraic combinations of them.

---

## What an Epsilon-Delta Proof Is Really Saying

When we write

$$\lim_{x \to a} f(x) = L,$$

we mean: every time someone demands that the output be within a tiny distance $\epsilon$ of $L$, we can respond by choosing a tiny distance $\delta$ so that keeping $x$ within $\delta$ of $a$ guarantees the output stays within $\epsilon$ of $L$.

The roles are different:

- $\epsilon$ is the challenge.
- $\delta$ is our answer to that challenge.
- $0 < |x-a| < \delta$ means: stay close to $a$, but do not worry about plugging in exactly $a$.
- $|f(x)-L| < \epsilon$ means: the output error is smaller than the demanded tolerance.

A good mental model is this: $\epsilon$ controls vertical error, $\delta$ controls horizontal error, and the proof explains how much horizontal control is enough to guarantee the desired vertical control.

The strategy is almost always the same:

1. Start from the output condition you want: $|f(x)-L| < \epsilon$.
2. Rewrite or bound it until it depends on $|x-a|$.
3. Choose $\delta$ small enough to make that final bound less than $\epsilon$.

## Example 1: Why $\delta = \epsilon/2$ for $\lim_{x \to 4}(2x-3)=5$

We want to prove:

$$\lim_{x \to 4}(2x-3) = 5.$$

### Step 1: Write the output error

$$|(2x-3)-5| = |2x-8| = 2|x-4|.$$

This is the whole story. The output error is exactly twice the input error.

### Step 2: Connect this to $\delta$

If $|x-4| < \delta$, then

$$|(2x-3)-5| = 2|x-4| < 2\delta.$$

So the output error is guaranteed to be less than $2\delta$.

### Step 3: Why choose $\delta = \epsilon/2$?

Our target is to make the output error smaller than $\epsilon$.

From Step 2, we already know the output error is less than $2\delta$. So it is enough to make $2\delta$ equal to $\epsilon$ or smaller.

That means we want

$$2\delta \leq \epsilon.$$

Now solve for $\delta$ by dividing by $2$:

$$\delta \leq \epsilon/2.$$

So any choice with $\delta \leq \epsilon/2$ will work. The simplest choice is

$$\delta = \epsilon/2.$$

Then the logic is:

$$|x-4| < \delta = \epsilon/2$$

so

$$|(2x-3)-5| = 2|x-4| < 2(\epsilon/2) = \epsilon.$$

The strict inequality comes from $|x-4| < \delta$, not from $2\delta < \epsilon$. That is the part that usually feels slippery the first time.

### Intuition

The function $2x-3$ doubles horizontal mistakes. If $x$ is off by $0.01$, the output is off by $0.02$. So to keep the output within $\epsilon$, we cut the allowed input error in half. That is why $\delta = \epsilon/2$ is not magic; it is just solving the inequality created by the function's scaling behavior.

### Concrete example

If someone demands $\epsilon = 0.1$, then we choose $\delta = 0.05$. Keeping $x$ within $0.05$ of $4$ guarantees the output stays within $0.1$ of $5$.

## Intuitive First-Principles Breakdown of $$\lim_{x \to 5} \frac{1}{x-3} = \frac{1}{2}$$

Explain to me intuitively from first principle, epsilon-delta proofs.

Now:

$$\lim_{x \to 5} \frac{1}{x-3} = \frac{1}{2}$$

The denominator is $x-3$, not $2$.

Break it down. Do not assume.

To understand epsilon-delta ($\epsilon$-$\delta$) proofs intuitively from first principles, we have to look at what a limit actually is: a game of precision and control.

When we say

$$\lim_{x \to 5} \frac{1}{x-3} = \frac{1}{2},$$

we are making a strict promise:

"I can get the output $\frac{1}{x-3}$ as close to $\frac{1}{2}$ as you could ever want, within a tiny target error $\epsilon$, just by forcing the input $x$ to be close enough to $5$, within a tiny distance $\delta$."

### The First-Principles Intuition

Think of $\epsilon$ as the tolerance challenge someone gives you. They say, "Make the error on the vertical axis smaller than $\epsilon$."

Think of $\delta$ as your control knob on the horizontal axis. You have to find a setting on your knob so that as long as $x$ stays within $\delta$ of $5$, the output is guaranteed to fall inside the target zone.

Mathematically, we write this relationship as:

If $$0 < |x-5| < \delta,$$ then

$$\left|\frac{1}{x-3} - \frac{1}{2}\right| < \epsilon.$$

Our entire goal in a proof is to look at the error we want to bound and reverse-engineer it to figure out how tight our control knob $\delta$ needs to be.

## Breaking Down the Proof Step by Step

Let us prove

$$\lim_{x \to 5} \frac{1}{x-3} = \frac{1}{2}$$

without skipping a single step.

### Step 1: Analyze the Error

We start with the output error expression and simplify it algebraically to see how it connects to the input distance $|x-5|$:

$$\left|\frac{1}{x-3} - \frac{1}{2}\right|.$$

Find a common denominator inside the absolute value:

$$\left|\frac{2-(x-3)}{2(x-3)}\right|$$

$$= \left|\frac{2-x+3}{2(x-3)}\right|$$

$$= \left|\frac{5-x}{2(x-3)}\right|$$

Because $|5-x|$ is exactly the same distance as $|x-5|$, we can rewrite this as:

$$\left|\frac{1}{x-3} - \frac{1}{2}\right| = \frac{|x-5|}{2|x-3|}.$$

This is the key object.

Right away you can see the difficulty: the error is not just a constant times $|x-5|$. The denominator still depends on $x$.

### Step 2: The Trap, Handling the Wild Variable

Here is the problem: we want our control knob $\delta$ to be a clean constant depending only on $\epsilon$. But right now, the bound still depends on $|x-3|$, which changes with $x$.

We need to put a safety net around $x$ so that the denominator $|x-3|$ does not do anything dangerous, like shrinking toward $0$ and making the fraction blow up.

Let us establish a standard preliminary restriction:

$$\delta \leq 1.$$

If $x$ is within a distance of $1$ from $5$, then

$$4 < x < 6.$$

Now subtract $3$ from all parts of the inequality:

$$1 < x-3 < 3.$$

Because $x-3$ is positive in this interval, we get

$$1 < |x-3| < 3.$$

So inside this safe zone, the denominator can no longer get close to $0$.

### Step 3: Finding the Worst-Case Scenario

Look back at the error from Step 1:

$$\frac{|x-5|}{2|x-3|}.$$

To make sure this fraction stays smaller than $\epsilon$, we look at the worst-case scenario for the denominator. A fraction gets larger when its denominator gets smaller.

From the safety-net step, the smallest $|x-3|$ can possibly get is $1$.

So if $|x-3| > 1$, then it must be true that

$$\frac{1}{|x-3|} < \frac{1}{1} = 1.$$

Substituting this worst-case scenario back into the expression gives

$$\frac{|x-5|}{2|x-3|} < \frac{|x-5|}{2 \cdot 1} = \frac{|x-5|}{2}.$$

If we can force this worst-case ceiling to be less than $\epsilon$, then the original fraction is guaranteed to be less than $\epsilon$ too.

### Step 4: Setting the Control Knob

Now we solve the simpler inequality:

$$\frac{|x-5|}{2} < \epsilon.$$

Multiply both sides by $2$:

$$|x-5| < 2\epsilon.$$

So we now have two constraints on our control knob $\delta$:

- $\delta \leq 1$ keeps the denominator safe.
- $\delta \leq 2\epsilon$ makes the error small enough.

To satisfy both at the same time, we choose

$$\delta = \min(1, 2\epsilon).$$

## Writing the Formal Proof

Given any $\epsilon > 0$, let

$$\delta = \min(1, 2\epsilon).$$

Suppose

$$0 < |x-5| < \delta.$$

Because $\delta \leq 1$, we know $|x-5| < 1$, so $4 < x < 6$, and therefore $1 < |x-3| < 3$. In particular, $|x-3| > 1$, which means

$$\frac{1}{|x-3|} < 1.$$

Because $\delta \leq 2\epsilon$, we also know that

$$|x-5| < 2\epsilon.$$

Now evaluate the distance between the function and the limit:

$$\left|\frac{1}{x-3} - \frac{1}{2}\right| = \frac{|x-5|}{2|x-3|}.$$

Using the two facts we established, we get

$$\frac{|x-5|}{2|x-3|} < \frac{2\epsilon}{2 \cdot 1} = \epsilon.$$

Therefore

$$\left|\frac{1}{x-3} - \frac{1}{2}\right| < \epsilon,$$

which completes the proof.

## Follow-Up: Lost at Step 3

### Step 3: Finding the Worst-Case Scenario

Look back at our goal from Step 1:

$$\frac{|x - 5|}{2|x - 3|} < \epsilon.$$

To make sure this fraction stays smaller than $\epsilon$, we must look at the worst-case scenario for the denominator. The fraction becomes largest when the denominator is at its smallest.

From our safety-net step, the smallest $|x - 3|$ can possibly get is $1$.

So, if $|x-3| > 1$, then it must be true that

$$\frac{1}{|x-3|} < \frac{1}{1}.$$

Substituting this worst-case scenario back into our expression:

$$\frac{|x - 5|}{2|x - 3|} < \frac{|x - 5|}{2 \cdot 1} = \frac{|x - 5|}{2}.$$

It is completely normal to get tripped up here. This is the exact moment where epsilon-delta proofs stop feeling like normal algebra and start feeling like a weird strategy game.

Let us slow down and look at why we do this, using a real-world analogy instead of just symbols.

### The Goal: Building a Safe Ceiling

Imagine you are designing a braking system for a car. You want to guarantee that the stopping distance is always less than $10$ meters, which plays the role of $\epsilon$.

To prove your system is safe, you do not test it under perfect conditions. You test it in the absolute worst-case scenario, like heavy rain and bald tires. If your car still stops in less than $10$ meters during the worst case, then it is automatically guaranteed to stop in less than $10$ meters during any normal day.

That is exactly what we are doing in Step 3.

We want to guarantee that our actual error expression is small:

$$\text{Actual Error} = \frac{|x-5|}{2|x-3|} < \epsilon.$$

Because $x$ is allowed to move around, the denominator $|x-3|$ keeps changing. To handle this, we construct a worst-case ceiling expression that is always bigger than our actual error.

If we can force that ceiling to stay under $\epsilon$, then our actual error is trapped underneath it and is guaranteed to be safe too.

### Why does the denominator get replaced by $1$?

Look at the fraction

$$\frac{|x-5|}{2|x-3|}.$$

If you want to make a fraction as large and dangerous as possible, do you make the denominator bigger or smaller?

$$\frac{10}{5} = 2, \quad \frac{10}{2} = 5, \quad \frac{10}{1} = 10.$$

Shrinking the denominator makes the whole fraction blow up.

From Step 2, our safety net forced $x$ to stay between $4$ and $6$. Because of that, the value of $|x-3|$ is trapped between $1$ and $3$.

The smallest the denominator can possibly get in our zone is $1$.

Therefore, the absolute largest, most dangerous value our error fraction could ever climb to is when we replace $|x-3|$ with its smallest possible value, namely $1$:

$$\text{Actual Error} = \frac{|x-5|}{2|x-3|} \leq \frac{|x-5|}{2 \cdot 1} = \text{Worst-Case Ceiling}.$$

### The Final Move

Now we have the chain of logic

$$\text{Actual Error} \leq \frac{|x-5|}{2}.$$

If we simply make sure that

$$\frac{|x-5|}{2} < \epsilon,$$

then by default the actual error must be less than $\epsilon$.

Multiplying both sides by $2$ gives

$$|x-5| < 2\epsilon.$$

This tells us exactly how to set our control knob. If we keep $x$ within $2\epsilon$ of $5$, even in the worst-case scenario where the denominator shrinks to $1$, the error still will not breach $\epsilon$.

## Follow-Up: Why is $2\epsilon$ There?

It is completely understandable to be confused by the $2\epsilon$. It feels weird to solve for something and end up with a number multiplied by $\epsilon$ instead of just $\epsilon$.

Let us look at why that $2$ is there, why it is completely allowed, and how it actually gives you more control.

### The Intuition: The Multiplier Effect

Think of the function like a mechanical steering system on a car, and think of the input $x$ as the steering wheel.

If you turn the steering wheel by $1$ degree, how much do the actual tires move? On some cars, turning the wheel a tiny bit makes the tires turn a lot. On other cars, you have to turn the wheel a lot just to get the tires to move a tiny bit.

In this problem, the function has a built-in dampening effect near $x=5$. Because of the denominator, any change you make to the input $x$ gets cut in half by the time it reaches the output.

If you want the output error to be less than $1$, you can actually afford to let $x$ wander by up to $2$ units. If you want the output error to be less than $0.01$, you can let $x$ wander by $0.02$ units.

Because the function naturally dampens your input movements, your control knob $\delta$ can be twice as wide as the target error $\epsilon$. That is why $\delta = 2\epsilon$ appears in the bound.

### The Math: Walking It Backward

We discovered that our worst-case error looks like this:

$$\text{Error} = \frac{|x-5|}{2}.$$

Remember, our ultimate goal is to force this error to be smaller than $\epsilon$:

$$\frac{|x-5|}{2} < \epsilon.$$

To isolate $|x-5|$, multiply both sides by $2$:

$$|x-5| < 2\epsilon.$$

This is telling you a direct rule: as long as the distance between $x$ and $5$ is smaller than $2\epsilon$, your error will be smaller than $\epsilon$.

### Test It with Real Numbers

Suppose someone challenges you with $\epsilon = 0.1$. They demand: make sure the output is within $0.1$ of $\frac{1}{2}$.

Using the rule above, you can set

$$\delta = 2\epsilon = 2(0.1) = 0.2.$$

This means you promise to keep $x$ within $0.2$ of $5$. Let us pick a number at the edge of that boundary, like $x = 5.2$.

Then

$$\frac{1}{5.2-3} = \frac{1}{2.2} \approx 0.4545.$$

Now compare this to the target $\frac{1}{2} = 0.5$:

$$\text{Actual Error} = |0.4545 - 0.5| = 0.0455.$$

This is smaller than the requested $\epsilon = 0.1$.

So the $2\epsilon$ means your control knob is allowed to be a little wider because the function's algebra naturally shrinks the error for you.

## What to Remember When You Get Stuck

- A limit claim is a promise: every output tolerance $\epsilon$ can be met by some input tolerance $\delta$.
- $\epsilon$ is given to you. $\delta$ is chosen by you.
- In linear examples, rewrite the output error until it is a constant times $|x-a|$.
- In rational examples, first keep the denominator away from $0$ by forcing $x$ into a safe interval.
- After that, bound the messy expression by something simpler.
- Then choose $\delta$ to make the simple expression smaller than $\epsilon$.

One sentence summary: an epsilon-delta proof is just a precise explanation of how much you need to control the input in order to guarantee a requested level of control on the output.
