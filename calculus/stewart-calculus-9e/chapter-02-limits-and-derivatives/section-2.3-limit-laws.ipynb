## The Limit Laws

Suppose the limits $\lim_{x \to a} f(x)$ and $\lim_{x \to a} g(x)$ both exist. Then:

| # | Law | Formula |
|---|-----|---------|
| 1 | **Sum Law** | $\lim_{x \to a}[f(x) + g(x)] = \lim_{x \to a} f(x) + \lim_{x \to a} g(x)$ |
| 2 | **Difference Law** | $\lim_{x \to a}[f(x) - g(x)] = \lim_{x \to a} f(x) - \lim_{x \to a} g(x)$ |
| 3 | **Constant Multiple Law** | $\lim_{x \to a}[c \cdot f(x)] = c \cdot \lim_{x \to a} f(x)$ |
| 4 | **Product Law** | $\lim_{x \to a}[f(x) \cdot g(x)] = \lim_{x \to a} f(x) \cdot \lim_{x \to a} g(x)$ |
| 5 | **Quotient Law** | $\lim_{x \to a}\frac{f(x)}{g(x)} = \frac{\lim_{x \to a} f(x)}{\lim_{x \to a} g(x)}$, provided $\lim_{x \to a} g(x) \neq 0$ |
| 6 | **Power Law** | $\lim_{x \to a}[f(x)]^n = \left[\lim_{x \to a} f(x)\right]^n$, where $n$ is a positive integer |
| 7 | **Root Law** | $\lim_{x \to a}\sqrt[n]{f(x)} = \sqrt[n]{\lim_{x \to a} f(x)}$, where $n$ is a positive integer |

These laws say: **you can break a limit apart algebraically**. Instead of computing the limit of a complicated expression all at once, you can split it into pieces, find each piece's limit separately, then reassemble.

## The Two Special Limits (Laws 8 & 9)

To actually *use* the seven laws, you need base cases — limits you can evaluate without any further calculation. There are exactly two:

$$\textbf{Law 8:} \quad \lim_{x \to a} c = c$$

$$\textbf{Law 9:} \quad \lim_{x \to a} x = a$$

### Why these feel like tricks (but aren't)

**Law 8 in plain English:** If your function just outputs the same constant $c$ no matter what $x$ is, then as $x$ approaches $a$, the output is still just $c$. There's no movement, no approach — it's always $c$.

**Law 9 in plain English:** If your function just outputs $x$ itself, then as $x$ approaches $a$, the output approaches $a$. Because the output *is* $x$.

These are so obvious that they feel like they shouldn't need to be stated — and that's exactly why the textbook says they're *"obvious from an intuitive point of view."* The graphs make this impossible to miss:

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import pandas as pd

%matplotlib inline
plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams['figure.figsize'] = [12, 5]
plt.rcParams['font.size'] = 12
```

```python
# Visualizing the two special limits: y = c and y = x

a = 2      # The point we're approaching
c = 3.5    # The constant

fig, axes = plt.subplots(1, 2, figsize=(13, 5))

x = np.linspace(-1, 5, 200)

# --- Left plot: y = c (constant function) ---
ax1 = axes[0]
ax1.plot(x, np.full_like(x, c), 'b-', linewidth=2.5, label=f'$y = c = {c}$')
ax1.axvline(x=a, color='gray', linestyle=':', linewidth=1.2, label=f'$x = a = {a}$')
ax1.axhline(y=c, color='red', linestyle='--', linewidth=1.2, alpha=0.7, label=f'$y = {c}$')

# Mark the point (a, c)
ax1.plot(a, c, 'ro', markersize=10, zorder=5)
ax1.annotate(f'$(a, c) = ({a}, {c})$', xy=(a, c), xytext=(2.5, 3.0),
             fontsize=10, arrowprops=dict(arrowstyle='->', color='red'))

# Arrows showing x approaching a
ax1.annotate('', xy=(a - 0.1, c - 0.4), xytext=(a - 1.5, c - 0.4),
             arrowprops=dict(arrowstyle='->', color='green', lw=1.5))
ax1.annotate('', xy=(a + 0.1, c - 0.4), xytext=(a + 1.5, c - 0.4),
             arrowprops=dict(arrowstyle='->', color='green', lw=1.5))
ax1.text(a, c - 0.7, '$x \\to a$ from both sides', ha='center', fontsize=9, color='green')
ax1.text(a, c + 0.3, 'output never moves', ha='center', fontsize=9, color='navy')

ax1.set_xlim(-0.5, 5)
ax1.set_ylim(0, 5)
ax1.set_xlabel('$x$')
ax1.set_ylabel('$y$')
ax1.set_title(r'Law 8: $\lim_{x \to a} c = c$', fontsize=13)
ax1.legend(loc='lower right', fontsize=9)

# --- Right plot: y = x (identity function) ---
ax2 = axes[1]
ax2.plot(x, x, 'b-', linewidth=2.5, label='$y = x$')
ax2.axvline(x=a, color='gray', linestyle=':', linewidth=1.2, label=f'$x = a = {a}$')
ax2.axhline(y=a, color='red', linestyle='--', linewidth=1.2, alpha=0.7, label=f'$y = a = {a}$')

# Mark the point (a, a)
ax2.plot(a, a, 'ro', markersize=10, zorder=5)
ax2.annotate(f'$(a, a) = ({a}, {a})$', xy=(a, a), xytext=(3.0, 1.2),
             fontsize=10, arrowprops=dict(arrowstyle='->', color='red'))

ax2.text(1.0, 3.5, 'output tracks input\nexactly (slope = 1)', ha='center', fontsize=9, color='navy')

ax2.set_xlim(-0.5, 5)
ax2.set_ylim(-0.5, 5)
ax2.set_xlabel('$x$')
ax2.set_ylabel('$y$')
ax2.set_title(r'Law 9: $\lim_{x \to a} x = a$', fontsize=13)
ax2.legend(loc='lower right', fontsize=9)

plt.suptitle('The Two Special Limits — No Calculation Required', fontsize=14, y=1.02)
plt.tight_layout()
plt.show()

print("Key observations:")
print("  y = c : The graph is a flat horizontal line. No matter where x goes, output = c. Always.")
print("  y = x : The diagonal line. As x creeps toward a, y creeps to exactly a. Output IS input.")
print("\nThese are the base cases that make all other limit laws computable.")
```

## Power Law for $x^n$ (Law 10)

From Law 9 and the Product Law (Law 4), we get another important special limit:

$$\lim_{x \to a} x^n = a^n \quad \text{where } n \text{ is a positive integer}$$

### Why this works — the chain argument

Since $\lim_{x \to a} x = a$, and the limit of a product equals the product of the limits (Law 4), you can just stack that rule $n$ times:

$$\lim_{x \to a} x^n = \lim_{x \to a}[\underbrace{x \cdot x \cdot \ldots \cdot x}_{n \text{ times}}] = \underbrace{a \cdot a \cdot \ldots \cdot a}_{n \text{ times}} = a^n$$

Every power function $x^n$ is just $x$ multiplied by itself $n$ times. Each factor contributes a limit of $a$. Chain it $n$ times and you land on $a^n$ — every time, no drama.

Visually, all curves $y = x^n$ pass **smoothly** through the point $(a, a^n)$ with no breaks or jumps. The function just *arrives* there continuously.

```python
# Visualizing lim x^n = a^n for various n

a = 2  # approaching x = 2
x = np.linspace(0, 3, 300)

fig, axes = plt.subplots(1, 4, figsize=(16, 4))
powers = [1, 2, 3, 4]
colors = ['steelblue', 'darkorange', 'green', 'purple']

for ax, n, color in zip(axes, powers, colors):
    y = x**n
    limit_val = a**n
    
    ax.plot(x, y, color=color, linewidth=2.5, label=f'$y = x^{n}$')
    ax.axvline(x=a, color='gray', linestyle=':', linewidth=1.2)
    ax.plot(a, limit_val, 'ro', markersize=10, zorder=5, label=f'$({a}, {limit_val})$')
    ax.axhline(y=limit_val, color='red', linestyle='--', alpha=0.5)
    
    ax.set_xlabel('$x$')
    ax.set_title(f'$\\lim_{{x \\to {a}}} x^{n} = {limit_val}$', fontsize=11)
    ax.set_xlim(0, 3)
    ax.set_ylim(0, min(3**n, 30))
    ax.legend(loc='upper left', fontsize=9)

plt.suptitle(r'Power Law: $\lim_{x \to a} x^n = a^n$ — The curve arrives smoothly at $(a, a^n)$',
             fontsize=13, y=1.02)
plt.tight_layout()
plt.show()

print("No breaks, no holes, no vertical asymptotes near x = 2.")
print("Every power function x^n is continuous at every real number a.")
print("The limit is just the function value — plug in and done.")
```

## Direct Substitution — When You Can Just Plug In

### The Big Shortcut

Using Laws 1–10, you can show that for **polynomials** and **rational functions**, limits can be evaluated by direct substitution — as long as the function is defined at the point:

$$\lim_{x \to a} p(x) = p(a) \quad \text{(for any polynomial } p\text{)}$$

$$\lim_{x \to a} \frac{p(x)}{q(x)} = \frac{p(a)}{q(a)} \quad \text{(provided } q(a) \neq 0\text{)}$$

**Why?** Because any polynomial is just sums and products of constants and $x$ — so you chain Laws 1–10 and end up with: just plug in $a$.

### Example 2(a): $\lim_{x \to 5} (2x^2 - 3x + 4)$

$$\lim_{x \to 5}(2x^2 - 3x + 4) = 2(5)^2 - 3(5) + 4 = 50 - 15 + 4 = 39$$

Notice $f(5) = 39$ — the limit equals the function value. Direct substitution works.

```python
# Direct substitution examples

def poly(x):
    return 2*x**2 - 3*x + 4

a = 5

print("Example 2(a): lim_{x→5} (2x² - 3x + 4)")
print("=" * 50)
print(f"\nDirect substitution: f(5) = 2(25) - 3(5) + 4 = {poly(5)}")
print()

# Verify numerically
x_left  = [4.9, 4.99, 4.999, 4.9999]
x_right = [5.1, 5.01, 5.001, 5.0001]

print(f"{'x':<12} {'f(x)':<15}")
print("-" * 27)
for x in x_left:
    print(f"{x:<12} {poly(x):<15.6f}")
print(f"{'→ 5 ←':<12} {'→ 39 ←':<15}")
for x in x_right:
    print(f"{x:<12} {poly(x):<15.6f}")

print("\nConclusion: lim = 39 = f(5). Direct substitution works perfectly.")
```

## The Indeterminate Form $\frac{0}{0}$ — When You Can't Just Plug In

When direct substitution gives $\frac{0}{0}$, this is called an **indeterminate form**. It's not that the limit doesn't exist — it's a **signal to factor and cancel** before substituting.

### Why the 0/0 signal works

The $\frac{0}{0}$ comes from a **cancellable common factor** — the same factor that makes both numerator and denominator zero. Once you cancel it, you're left with a simpler function that *is* continuous at $a$, so direct substitution works on that simpler version.

### Example 3: $\lim_{x \to 1} \frac{x^2 - 1}{x - 1}$

Direct substitution gives $\frac{0}{0}$. Factor:

$$\frac{x^2 - 1}{x - 1} = \frac{(x-1)(x+1)}{x-1} = x + 1 \quad \text{(for } x \neq 1\text{)}$$

$$\lim_{x \to 1} \frac{x^2 - 1}{x - 1} = \lim_{x \to 1} (x + 1) = 2$$

### The Deep Insight

The key principle at work here:

> **If $f(x) = g(x)$ when $x \neq a$, then $\lim_{x \to a} f(x) = \lim_{x \to a} g(x)$, provided the limits exist.**

This is the freedom that limits grant you. Since we **never** need the value at $x = a$ — only what happens *approaching* $a$ — we are free to assume $x \neq a$ the entire time we simplify. And under that assumption, any cancellation is perfectly legal.

We're not removing $(x-1)$ because it's zero. We're removing it because **we're only looking at $x \neq 1$**, and for those values, it's a perfectly legal cancellation. The limit doesn't know or care about the swap — it never looked at $x = a$ in the first place.

```python
# Example 3: lim (x^2 - 1)/(x - 1) as x -> 1
# Demonstrating the 0/0 case and the factoring trick

def f_original(x):
    """f(x) = (x^2 - 1)/(x - 1) — undefined at x = 1"""
    return (x**2 - 1) / (x - 1)

def g_simplified(x):
    """g(x) = x + 1 — the simplified version, valid for all x"""
    return x + 1

# Show numerical values approaching 1
x_vals = [0.5, 0.9, 0.99, 0.999, 0.9999, 1.0001, 1.001, 1.01, 1.1, 1.5]

print("lim_{x→1} (x²-1)/(x-1)")
print("=" * 50)
print(f"\n{'x':<10} {'f(x) original':<20} {'g(x) = x+1':<15}")
print("-" * 45)

for x in x_vals:
    f_val = f_original(x)
    g_val = g_simplified(x)
    print(f"{x:<10} {f_val:<20.6f} {g_val:<15.6f}")

print()
print("At x = 1: f(1) = 0/0 (undefined), but g(1) = 2")
print("f and g are IDENTICAL for every x ≠ 1")
print("Therefore: lim_{x→1} f(x) = lim_{x→1} g(x) = g(1) = 2")
```

```python
# Visualizing Example 3: the hole in f(x) vs the solid point in g(x)

fig, axes = plt.subplots(1, 2, figsize=(13, 5))

x_plot = np.linspace(-0.5, 3, 400)
x_no_1 = x_plot[np.abs(x_plot - 1) > 0.01]  # Exclude near x=1

# --- f(x) = (x^2-1)/(x-1) ---
ax1 = axes[0]
ax1.plot(x_no_1, f_original(x_no_1), 'b-', linewidth=2.5, label=r'$f(x) = \frac{x^2-1}{x-1}$')
# Open circle at (1, 2) — hole
ax1.plot(1, 2, 'wo', markersize=12, zorder=5, markeredgecolor='blue', markeredgewidth=2)
ax1.axvline(x=1, color='gray', linestyle=':', linewidth=1)
ax1.axhline(y=2, color='red', linestyle='--', alpha=0.5, linewidth=1.2, label='$y = 2$')

ax1.annotate('Hole at (1, 2)\nf(1) undefined', xy=(1, 2), xytext=(1.5, 1.3),
             fontsize=10, color='red', arrowprops=dict(arrowstyle='->', color='red'))

ax1.set_xlim(-0.5, 3)
ax1.set_ylim(-0.5, 4)
ax1.set_xlabel('$x$')
ax1.set_ylabel('$y$')
ax1.set_title(r'$f(x) = \frac{x^2-1}{x-1}$ — has a hole at $x=1$', fontsize=11)
ax1.legend(loc='upper left', fontsize=10)

# --- g(x) = x + 1 ---
ax2 = axes[1]
ax2.plot(x_plot, g_simplified(x_plot), 'g-', linewidth=2.5, label='$g(x) = x + 1$')
ax2.plot(1, 2, 'go', markersize=12, zorder=5, label='$(1, 2)$ — defined!')
ax2.axvline(x=1, color='gray', linestyle=':', linewidth=1)
ax2.axhline(y=2, color='red', linestyle='--', alpha=0.5, linewidth=1.2, label='$y = 2$')

ax2.annotate('Solid point (1, 2)\ng(1) = 2', xy=(1, 2), xytext=(1.5, 1.3),
             fontsize=10, color='green', arrowprops=dict(arrowstyle='->', color='green'))

ax2.set_xlim(-0.5, 3)
ax2.set_ylim(-0.5, 4)
ax2.set_xlabel('$x$')
ax2.set_ylabel('$y$')
ax2.set_title(r'$g(x) = x+1$ — identical to $f$ everywhere except $x=1$', fontsize=11)
ax2.legend(loc='upper left', fontsize=10)

plt.suptitle('f and g have the same limit at x=1 because they agree everywhere x ≠ 1',
             fontsize=12, y=1.02)
plt.tight_layout()
plt.show()

print("The graphs are visually identical — the only difference is the single point at x = 1.")
print("A limit doesn't care about that single point. So both limits equal 2.")
```

## Example 5: $\lim_{h \to 0} \frac{(3+h)^2 - 9}{h}$

Direct substitution gives $\frac{0}{0}$. Expand and simplify:

$$\frac{(3+h)^2 - 9}{h} = \frac{9 + 6h + h^2 - 9}{h} = \frac{6h + h^2}{h} = \frac{h(6 + h)}{h} = 6 + h \quad (h \neq 0)$$

$$\lim_{h \to 0} \frac{(3+h)^2 - 9}{h} = \lim_{h \to 0}(6 + h) = 6$$

### The insight about canceling $h$

You might think: *"We removed $h$ because it was zero and we can't divide by zero."* That's true but it misses the deeper point.

The real reason we can cancel $h$ is the **freedom the limit grants us**: since we're computing a limit *as $h \to 0$*, we only ever consider $h \neq 0$. We never look at $h = 0$ itself. And if $h \neq 0$, then $h/h = 1$ — perfectly legal.

The cancellation happens **because of** the limit, not **despite** it.

The final substitution ($h = 0$ into $6 + h$) isn't plugging into the original forbidden function. It's plugging into the new, simplified function $(6 + h)$, which **is** defined at $h = 0$. You never broke any rules.

```python
# Example 5: lim (3+h)^2 - 9) / h as h -> 0

def F(h):
    """F(h) = ((3+h)^2 - 9) / h — undefined at h = 0"""
    return ((3 + h)**2 - 9) / h

def G(h):
    """G(h) = 6 + h — simplified, valid for all h"""
    return 6 + h

h_vals = [1.0, 0.5, 0.1, 0.05, 0.01, 0.001, 0.0001,
         -0.0001, -0.001, -0.01, -0.05, -0.1, -0.5, -1.0]

print("lim_{h→0} ((3+h)² - 9) / h")
print("=" * 45)
print(f"\n{'h':<12} {'F(h) original':<20} {'G(h) = 6+h':<15}")
print("-" * 47)

for h in sorted(h_vals):
    print(f"{h:<12.4f} {F(h):<20.8f} {G(h):<15.8f}")

print()
print("As h → 0 (from both sides), F(h) → 6")
print("G(0) = 6 + 0 = 6 ✓")
print()
print("F(h) = G(h) for all h ≠ 0, so their limits are equal.")
print("The limit is 6 — which is the instantaneous rate of change of x² at x = 3.")
```

```python
# Visualize Example 5

fig, axes = plt.subplots(1, 2, figsize=(13, 5))

h_plot = np.linspace(-2, 2, 400)
h_no_0 = h_plot[np.abs(h_plot) > 0.02]

# --- F(h) --- 
ax1 = axes[0]
ax1.plot(h_no_0, F(h_no_0), 'b-', linewidth=2.5, label=r'$F(h) = \frac{(3+h)^2-9}{h}$')
ax1.plot(0, 6, 'wo', markersize=12, zorder=5, markeredgecolor='blue', markeredgewidth=2)
ax1.axhline(y=6, color='red', linestyle='--', alpha=0.5, linewidth=1.2, label='$y = 6$')
ax1.axvline(x=0, color='gray', linestyle=':', linewidth=1)
ax1.annotate('Hole at (0, 6)\nF(0) undefined', xy=(0, 6), xytext=(0.5, 5.0),
             fontsize=10, color='red', arrowprops=dict(arrowstyle='->', color='red'))
ax1.set_xlim(-2, 2)
ax1.set_ylim(2, 10)
ax1.set_xlabel('$h$')
ax1.set_ylabel('value')
ax1.set_title(r'$F(h) = \frac{(3+h)^2-9}{h}$ — hole at $h=0$', fontsize=11)
ax1.legend(fontsize=9)

# --- G(h) = 6 + h ---
ax2 = axes[1]
ax2.plot(h_plot, G(h_plot), 'g-', linewidth=2.5, label='$G(h) = 6 + h$')
ax2.plot(0, 6, 'go', markersize=12, zorder=5, label='$(0, 6)$ — defined!')
ax2.axhline(y=6, color='red', linestyle='--', alpha=0.5, linewidth=1.2, label='$y = 6$')
ax2.axvline(x=0, color='gray', linestyle=':', linewidth=1)
ax2.set_xlim(-2, 2)
ax2.set_ylim(2, 10)
ax2.set_xlabel('$h$')
ax2.set_ylabel('value')
ax2.set_title(r'$G(h) = 6+h$ — continuous, $G(0)=6$', fontsize=11)
ax2.legend(fontsize=9)

plt.suptitle('Example 5: Canceling h is legal because the limit never uses h = 0',
             fontsize=12, y=1.02)
plt.tight_layout()
plt.show()
```

## The Core Principle: Swapping to a Simpler Twin

The general rule that justifies Examples 3–5:

> **If $f(x) = g(x)$ whenever $x \neq a$, then**
> $$\lim_{x \to a} f(x) = \lim_{x \to a} g(x)$$
> **(provided the limits exist)**

This is what makes algebraic manipulation inside limits so powerful. You're not manipulating the limit itself — you're **swapping the function for a simpler identical twin** that happens to be well-behaved at the point you care about.

The limit doesn't know or care about the swap, because it **never looked at $x = a$ in the first place**.

### Checklist for computing limits when direct substitution gives $\frac{0}{0}$:

1. **Factor** both numerator and denominator
2. **Cancel** the common factor (legal because we're assuming $x \neq a$)
3. **Substitute** directly into the simplified expression

```python
# More examples of the 0/0 form

print("Additional 0/0 limit examples:")
print("=" * 60)

# Example A: lim (x^2 - 4) / (x - 2) as x -> 2
print("\n(A) lim_{x→2} (x² - 4) / (x - 2)")
print("Direct sub: (4-4)/(2-2) = 0/0 → factor")
print("(x²-4)/(x-2) = (x-2)(x+2)/(x-2) = x+2  (x ≠ 2)")
print("lim = 2 + 2 = 4")
x_vals_A = [1.9, 1.99, 1.999, 2.001, 2.01, 2.1]
results_A = [(x**2 - 4) / (x - 2) for x in x_vals_A]
df_A = pd.DataFrame({'x': x_vals_A, 'f(x)': results_A})
print(df_A.to_string(index=False))

print()

# Example B: lim (x^3 - 8) / (x - 2) as x -> 2
print("(B) lim_{x→2} (x³ - 8) / (x - 2)")
print("Direct sub: (8-8)/(2-2) = 0/0 → factor using x³-a³ = (x-a)(x²+ax+a²)")
print("(x³-8)/(x-2) = (x-2)(x²+2x+4)/(x-2) = x²+2x+4  (x ≠ 2)")
print("lim = 4 + 4 + 4 = 12")
x_vals_B = [1.9, 1.99, 1.999, 2.001, 2.01, 2.1]
results_B = [(x**3 - 8) / (x - 2) for x in x_vals_B]
df_B = pd.DataFrame({'x': x_vals_B, 'f(x)': results_B})
print(df_B.to_string(index=False))

print()

# Example C: lim (sqrt(x) - 1) / (x - 1) as x -> 1
print("(C) lim_{x→1} (√x - 1) / (x - 1)")
print("Direct sub: 0/0 → rationalize: multiply by (√x + 1)/(√x + 1)")
print("(√x - 1)(√x + 1) / (x-1)(√x+1) = (x-1) / (x-1)(√x+1) = 1/(√x+1)")
print("lim = 1/(√1 + 1) = 1/2")
x_vals_C = [0.9, 0.99, 0.999, 1.001, 1.01, 1.1]
results_C = [(np.sqrt(x) - 1) / (x - 1) for x in x_vals_C]
df_C = pd.DataFrame({'x': x_vals_C, 'f(x)': results_C})
print(df_C.to_string(index=False))
```

## The Squeeze Theorem

If $f(x) \leq g(x) \leq h(x)$ near $a$, and $\lim_{x \to a} f(x) = \lim_{x \to a} h(x) = L$, then:

$$\lim_{x \to a} g(x) = L$$

The idea: if $g$ is always trapped between $f$ and $h$, and both $f$ and $h$ squeeze down to $L$ at $a$, then $g$ has no choice but to go to $L$ as well.

### Classic application: $\lim_{x \to 0} x^2 \sin\left(\frac{1}{x}\right)$

We know $-1 \leq \sin\left(\frac{1}{x}\right) \leq 1$ for all $x \neq 0$. Multiply through by $x^2 \geq 0$:

$$-x^2 \leq x^2 \sin\left(\frac{1}{x}\right) \leq x^2$$

Since $\lim_{x \to 0}(-x^2) = 0$ and $\lim_{x \to 0}(x^2) = 0$, by the Squeeze Theorem:

$$\lim_{x \to 0} x^2 \sin\left(\frac{1}{x}\right) = 0$$

```python
# Visualizing the Squeeze Theorem

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# --- x^2 sin(1/x) squeezed between -x^2 and x^2 ---
ax1 = axes[0]

x = np.linspace(-0.5, 0.5, 2000)
x_nz = x[np.abs(x) > 1e-10]  # Avoid division by zero

y_func  = x_nz**2 * np.sin(1 / x_nz)
y_upper = x_nz**2
y_lower = -x_nz**2

ax1.fill_between(x_nz, y_lower, y_upper, alpha=0.15, color='gray', label='Squeeze region')
ax1.plot(x_nz, y_upper, 'r--', linewidth=1.5, label='$x^2$')
ax1.plot(x_nz, y_lower, 'b--', linewidth=1.5, label='$-x^2$')
ax1.plot(x_nz, y_func, 'g-', linewidth=1.5, alpha=0.8, label='$x^2 \\sin(1/x)$')
ax1.plot(0, 0, 'ko', markersize=8, zorder=5, label='limit = 0')

ax1.set_xlim(-0.5, 0.5)
ax1.set_ylim(-0.3, 0.3)
ax1.set_xlabel('$x$')
ax1.set_ylabel('$y$')
ax1.set_title(r'$\lim_{x \to 0} x^2 \sin(1/x) = 0$ (Squeeze Theorem)', fontsize=11)
ax1.legend(fontsize=9)

# --- Zoom in near 0 ---
ax2 = axes[1]

x_zoom = np.linspace(-0.1, 0.1, 2000)
x_zoom_nz = x_zoom[np.abs(x_zoom) > 1e-10]

ax2.fill_between(x_zoom_nz, -x_zoom_nz**2, x_zoom_nz**2, alpha=0.15, color='gray')
ax2.plot(x_zoom_nz, x_zoom_nz**2, 'r--', linewidth=1.5, label='$x^2$ (upper bound)')
ax2.plot(x_zoom_nz, -x_zoom_nz**2, 'b--', linewidth=1.5, label='$-x^2$ (lower bound)')
ax2.plot(x_zoom_nz, x_zoom_nz**2 * np.sin(1 / x_zoom_nz), 'g-', linewidth=1, alpha=0.9,
         label='$x^2\\sin(1/x)$')
ax2.plot(0, 0, 'ko', markersize=10, zorder=5, label='squeezed to 0')

ax2.set_xlim(-0.1, 0.1)
ax2.set_ylim(-0.015, 0.015)
ax2.set_xlabel('$x$')
ax2.set_title('Zoomed in near $x = 0$', fontsize=11)
ax2.legend(fontsize=9)

plt.suptitle('Squeeze Theorem: $g$ is trapped → $g$ must go to the same limit', fontsize=13, y=1.02)
plt.tight_layout()
plt.show()
```

## Summary: The Full Toolkit for Section 2.3

| Situation | Strategy |
|-----------|----------|
| Function is a polynomial or rational, $a$ is in domain | **Direct substitution** — just plug in |
| Substitution gives $\frac{0}{0}$ | **Factor and cancel** the common factor |
| Has a radical and gives $\frac{0}{0}$ | **Rationalize** (multiply by conjugate) |
| Function is bounded between two known limits | **Squeeze Theorem** |
| One-sided behavior differs | **One-sided limits** separately |

### The overarching principle

A limit only cares about what happens **near** $a$, never **at** $a$. This freedom — to assume $x \neq a$ throughout all your algebra — is what makes every technique above valid. You're always swapping for an identical twin function that just happens to be better-behaved at the point you're approaching.
