## Introduction

Basic concepts:
- **Point** - has position but no size
- **Line** - extends infinitely in both directions
- **Segment** - portion of a line between two points
- **Ray** - line extending from a point in one direction

These are the building blocks for everything that follows.

## Chapter 1: The Straight Line

### §1 Angles

An **angle** is formed by two rays sharing a common endpoint (the vertex).

Key concepts:
- **Right angle** = 90° (quarter turn)
- **Straight angle** = 180° (half turn)
- **Supplementary angles** - two angles that sum to 180°
- **Complementary angles** - two angles that sum to 90°

---

### §2 Perpendicular Lines

Two lines are **perpendicular** if they intersect at a right angle (90°).

Notation: $AB \perp CD$ means line AB is perpendicular to line CD.

---

## Key Theorem: Uniqueness of Perpendicular from External Point

**Proposition 24:** From any point M lying outside a given line AB, one can drop a perpendicular to this line, and such perpendicular is **unique**.

### The Proof (Using Folding)

1. Given: Line AB and point M not on the line
2. Fold the diagram so the upper part meets the lower part
3. Point M moves to position N (its mirror image across AB)
4. Let C be where line MN crosses AB

**Claim:** MN ⊥ AB, and this is the only perpendicular from M.

**Why MN is perpendicular:**
- After folding, M lands on N
- C stays in place (it's on the fold line AB)
- So MC folds onto NC
- Therefore ∠MCB = ∠NCB (the fold preserves angles)
- But ∠MCB and ∠NCB are supplementary (they form a straight line MN)
- Two equal supplementary angles must each be 90°
- Therefore MN ⊥ AB ✓

**Why no other line through M is perpendicular:**

Consider any other line MD (where D ≠ C):
- After folding, M → N but D stays in place
- MD folds onto ND
- So ∠MDC = ∠NDC
- But MDN is NOT a straight line (because there's only one straight line from M to N)
- So ∠MDC + ∠NDC ≠ 180°
- Therefore ∠MDC ≠ 90°
- Therefore MD is NOT perpendicular to AB ✓

### My Intuition: The "Path to a Friend" Analogy

I like to think about this in terms of walking to a friend's house:

**Setup:**
- You're at point M
- Your friend is at point N (directly across the line AB)
- The line AB is a road you must cross

**The perpendicular (through C):**

If the crossing point C is the midpoint between M and N, then you walk in a **straight line** from M to N. No turning required. The angle at C is zero degrees of "turn" — which means you're crossing at 90° to the road.

**Any other path (through D):**

If you cross at some other point D (not the midpoint), your path MDN is NOT straight. You have to **turn** at D to reach N. This turning angle is greater than zero, which means you're NOT crossing perpendicular to the road.

**The key insight:**

There's only ONE straight line from M to N (basic axiom: two points determine a unique line). This line crosses AB at exactly one point C. That point C is where the perpendicular drops.

Any other crossing point D requires a "bent" path → not perpendicular.

### Visual Summary

```
         M
         |
         | (perpendicular - straight path)
         |
----A----C----B---- (the line)
         |
         |
         N

vs.

         M
          \
           \  (bent path - NOT perpendicular)
            \
----A----D----B----
           /
          /
         N
```

The straight path M→C→N is the perpendicular. Any bent path M→D→N is not.
