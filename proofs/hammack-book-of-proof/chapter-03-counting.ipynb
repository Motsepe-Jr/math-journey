This chapter is about counting the number of ways to arrange or select things. The main objects of study are **lists** (ordered sequences of symbols).

A list of length $k$ drawn from a set of $n$ symbols looks like:

$$(\_, \_, \_, \ldots, \_) \quad \leftarrow k \text{ slots}$$

Two key distinctions:
- **Repetition allowed** — the same symbol can appear more than once
- **Repetition not allowed** (non-repetitive list) — each symbol can appear at most once

---
## Lists with Repetition Allowed

**Rule:** To form a list of length $k$ from a set of $n$ symbols where repetition is allowed, there are $n$ choices for each slot:

$$\underbrace{n \cdot n \cdot n \cdots n}_{k \text{ times}} = n^k$$

**Example 3.2(a):** Lists of length 4 from $\{A, B, C, D, E, F, G\}$ (7 symbols), repetition allowed:

$$7 \cdot 7 \cdot 7 \cdot 7 = 7^4 = 2401$$

```python
n = 7
k = 4
total_with_rep = n**k
print(f"Lists of length {k} from {n} symbols (repetition allowed): {n}^{k} = {total_with_rep}")
```

---
## Lists Without Repetition (Non-Repetitive Lists)

**Rule:** To form a list of length $k$ from $n$ symbols where repetition is **not** allowed, each slot has one fewer choice than the previous:

$$n \cdot (n-1) \cdot (n-2) \cdots (n-k+1)$$

This is because once a symbol is placed in a slot, it can no longer be used in later slots.

**Example 3.2(b):** Lists of length 4 from 7 symbols, **no repetition**:

$$7 \cdot 6 \cdot 5 \cdot 4 = 840$$

Visualizing the slots:

| Slot 1 | Slot 2 | Slot 3 | Slot 4 |
|--------|--------|--------|--------|
| 7 choices | 6 choices | 5 choices | 4 choices |

```python
from math import prod

n, k = 7, 4

# No repetition: n * (n-1) * (n-2) * ... * (n-k+1)
no_rep = prod(range(n, n - k, -1))
print(f"Lists of length {k} from {n} symbols (no repetition): {' * '.join(str(x) for x in range(n, n-k, -1))} = {no_rep}")
```

## Complement Counting (Divide and Conquer)

> **Key idea:** Sometimes it is easier to count everything **except** what you want, and then subtract.
>
> $$\text{count}(\text{desired}) = \text{count}(\text{total}) - \text{count}(\text{not desired})$$

This is especially useful when the "not desired" case is simpler to count than the "desired" case directly.

---

### Example 3.2(c): No repetition, list must contain E

**Direct approach (by cases):** E must appear in exactly one of the 4 slots. Divide into 4 types based on where E lands:

| Type | Slot 1 | Slot 2 | Slot 3 | Slot 4 |
|------|--------|--------|--------|--------|
| Type 1 | **E** | 6 choices | 5 choices | 4 choices |
| Type 2 | 6 choices | **E** | 5 choices | 4 choices |
| Type 3 | 6 choices | 5 choices | **E** | 4 choices |
| Type 4 | 6 choices | 5 choices | 4 choices | **E** |

Each type gives $6 \cdot 5 \cdot 4 = 120$ lists. Total:

$$4 \times 120 = 480$$

---

### Example 3.2(d): Repetition allowed, list must contain E

**Complement approach:** Counting directly is hard because E could appear 1, 2, 3, or 4 times. Instead:

$$\text{lists with at least one E} = \underbrace{7^4}_{\text{total}} - \underbrace{6^4}_{\text{lists with NO E}}$$

$$= 2401 - 1296 = 1105$$

The "$6^4$ lists with no E" are built by choosing freely from the 6 symbols $\{A,B,C,D,F,G\}$ — E is simply excluded from all choices.

> **Why is complement easier here?** Because "at least one E" includes many sub-cases (E once, E twice, E three times, E four times), but "no E at all" is just one clean case.

```python
n = 7       # total symbols
k = 4       # list length
# No repetition, must contain E — direct by cases
e_no_rep = 4 * prod(range(n - 1, n - 1 - (k - 1), -1))   # 4 * 6*5*4
print(f"(c) No repetition, must contain E: 4 × {6}×{5}×{4} = {e_no_rep}")

# Repetition allowed, must contain E — complement
total = n**k
no_e  = (n - 1)**k
e_with_rep = total - no_e
print(f"\n(d) Repetition allowed, must contain E:")
print(f"    total={total}, no-E={no_e}, answer={e_with_rep}")
```

## Factorials

The **factorial** of a non-negative integer $n$, written $n!$, is the product of all positive integers up to $n$:

$$n! = n \cdot (n-1) \cdot (n-2) \cdots 2 \cdot 1$$

Special case: $0! = 1$ (by convention, and it makes formulas work out cleanly).

| $n$ | $n!$ |
|-----|------|
| 0 | 1 |
| 1 | 1 |
| 2 | 2 |
| 3 | 6 |
| 4 | 24 |
| 5 | 120 |
| 6 | 720 |
| 7 | 5040 |

**Connection to lists:** The number of non-repetitive lists of length $k$ from $n$ symbols can be written neatly using factorials:

$$n \cdot (n-1) \cdots (n-k+1) = \frac{n!}{(n-k)!}$$

For $k = n$ (arrange **all** symbols), this becomes $n!$ — the number of ways to arrange $n$ distinct objects in a row.

```python
from math import factorial

# Print factorial table
print(f"{'n':>4}  {'n!':>10}")
print("-" * 18)
for n in range(11):
    print(f"{n:>4}  {factorial(n):>10}")

# Verify: n!/(n-k)! = n*(n-1)*...*(n-k+1)
n, k = 7, 4
via_factorial = factorial(n) // factorial(n - k)
via_product   = prod(range(n, n - k, -1))
print(f"\n7! / (7-4)! = {factorial(7)} / {factorial(3)} = {via_factorial}")
print(f"7×6×5×4    = {via_product}")
print(f"Equal? {via_factorial == via_product}")
```
