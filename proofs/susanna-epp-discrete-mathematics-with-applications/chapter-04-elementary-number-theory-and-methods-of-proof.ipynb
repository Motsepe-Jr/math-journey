## Chapter 4: Elementary Number Theory and Methods of Proof

These are my study notes after working through Chapter 4. The chapter is partly about number systems and partly about how mathematical certainty is built through proof.

## Big Picture

The chapter works with integers, rational numbers, and real numbers, but its real theme is stronger than that: how to decide whether a mathematical statement is actually true or false.

A useful opening example uses the floor function $\lfloor x \rfloor$, which means the greatest integer less than or equal to $x$.

- $\lfloor 2.3 \rfloor = 2$
- $\lfloor 12.99999 \rfloor = 12$
- $\lfloor -1.5 \rfloor = -2$

Two warm-up questions capture the whole spirit of the chapter:

1. Is $\lfloor x + 1 \rfloor = \lfloor x \rfloor + 1$ true for every real number $x$?
2. Is $\lfloor x + y \rfloor = \lfloor x \rfloor + \lfloor y \rfloor$ true for every pair of real numbers $x$ and $y$?

The first is true and the second is false. The point is not just the answers. The point is how certain I am, and whether I can justify the answer instead of merely guessing it.

---

## What Proof Is Doing

A proof is a carefully reasoned argument aimed at a skeptical listener, which is often just myself reading my own work critically. At every step the listener can ask, "Why is that true?" A proof succeeds only if every step can answer that challenge.

One major lesson from the chapter is that intuition is not enough. A result can look obvious and still be false. Writing a proof forces hidden assumptions out into the open. That is one reason proof matters so much in mathematics and in applications: weak reasoning can stay hidden until the details are written down.

The chapter also stresses precision. In proof writing, saying exactly what I mean is not a style preference; it is part of the mathematics. Loose language usually signals loose reasoning.

## Assumptions and Core Definitions

The section assumes basic algebra, the standard properties of equality, substitution, and the fact that integers are closed under addition, subtraction, and multiplication. It also reminds me that quotients of integers are not always integers.

The most important definitions are:

- An integer $n$ is even if $n = 2k$ for some integer $k$.
- An integer $n$ is odd if $n = 2k + 1$ for some integer $k$.
- An integer $n > 1$ is prime if its only positive factors are $1$ and $n$.
- An integer $n > 1$ is composite if it can be written as a product of two smaller positive integers.

A key proof lesson is that definitions work in both directions when they are stated with if and only if.

- If I know an integer is even, then I may write it as $2k$.
- If I manage to write an integer as $2k$, then I may conclude it is even.

This is one of the main engines of direct proof.

## Main Proof Moves in This Section

### Constructive Existence

To prove an existential statement, I can produce one actual example or give a method for constructing one. This is a constructive proof of existence.

### Counterexample

To disprove a universal statement, I only need one counterexample. That means I find one case where the hypothesis is true but the conclusion fails.

### Exhaustion

If there are only finitely many relevant cases, I may check them all. This works in small settings, but it does not scale to infinite domains or to huge finite problems.

### Generalizing from the Generic Particular

This is the heart of direct proof. I pick an arbitrary element of the domain, assume the hypothesis holds for that element, and then prove the conclusion for that same element. Because the element was arbitrary, the argument applies to all elements.

### Direct Proof Template

1. Rewrite the claim mentally as: for every $x$ in a domain, if $P(x)$ then $Q(x)$.
2. Start with an arbitrary element satisfying the hypothesis.
3. Use definitions, algebra, and known facts to force the conclusion.

## Signature Example: Sum of Two Even Integers

The theorem says that the sum of any two even integers is even.

The direct-proof structure is clean:

- Let $m$ and $n$ be arbitrary even integers.
- Because they are even, write $m = 2r$ and $n = 2s$ for integers $r$ and $s$.
- Then $m + n = 2r + 2s = 2(r+s)$.
- Since integers are closed under addition, $r+s$ is an integer.
- Therefore $m+n$ has the form $2t$ for an integer $t$, so it is even.

This proof shows how definitions drive the whole argument. The proof is not a trick. It is just a disciplined use of the meaning of even.

## My Takeaways from Chapter 4

- Proof is about certainty, not plausibility.
- Definitions are not background vocabulary; they are tools used line by line in arguments.
- Counterexamples are often the fastest way to kill a false universal claim.
- Direct proof works by taking an arbitrary case and showing the conclusion must follow.
- Precision in language and notation is part of mathematical thinking itself.

This chapter feels like a transition from solving problems by instinct to solving them with controlled logical structure.
