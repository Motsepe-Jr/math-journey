### Section 1A: $\mathbb{R}^n$ and $\mathbb{C}^n$ — Exercises

---

Working through section 1A. This section is basically just establishing that $\mathbb{C}$ behaves the way you'd expect — commutativity, associativity, distributivity, inverses. Nothing surprising, but the point is to actually prove these things from the definitions rather than assume them.

The definition we're working from:

> **Complex addition:** $(a + bi) + (c + di) = (a+c) + (b+d)i$
>
> **Complex multiplication:** $(a + bi)(c + di) = (ac - bd) + (ad + bc)i$

where $a, b, c, d \in \mathbb{R}$.

---

## Exercise 1 — Commutativity of Addition

**Show that $\alpha + \beta = \beta + \alpha$ for all $\alpha, \beta \in \mathbb{C}$.**

---

### My Approach

The key here is that we know $\alpha$ and $\beta$ are complex numbers, so we can write them in terms of their real and imaginary parts and just use the definition of complex addition directly.

The reason this works is that commutativity of complex addition reduces to commutativity of real number addition — and we already know real numbers commute.

### Proof

Let $\alpha = a + bi$ and $\beta = c + di$ where $a, b, c, d \in \mathbb{R}$.

Then:

$$
\alpha + \beta = (a + bi) + (c + di)
$$

By the definition of complex addition:

$$
= (a + c) + (b + d)i
$$

Since $a, b, c, d \in \mathbb{R}$ and addition of real numbers is commutative:

$$
= (c + a) + (d + b)i
$$

Which by the definition of complex addition is just:

$$
= (c + di) + (a + bi) = \beta + \alpha
$$

Therefore $\alpha + \beta = \beta + \alpha$ for all $\alpha, \beta \in \mathbb{C}$. $\blacksquare$

---

### Notes

The thing I wrote in my notes was that this works because of **commutativity at the real level**. Once you expand using the definition of complex addition, the order of real numbers changes and that's it. The "grouping" (associativity) doesn't affect the result either — the point is just that $(a + c) = (c + a)$ and $(b + d) = (d + b)$ in $\mathbb{R}$.

## Exercise 2 — Associativity of Addition

**Show that $(\alpha + \beta) + \lambda = \alpha + (\beta + \lambda)$ for all $\alpha, \beta, \lambda \in \mathbb{C}$.**

---

### My Approach

Same idea as exercise 1 — write everything in terms of real and imaginary parts, apply the definition of complex addition, and let real number associativity do the work.

Associativity means grouping of numbers doesn't affect the result. I drew a diagram in my notes showing this.

### Proof

Let $\alpha = a + bi$, $\beta = c + di$, and $\lambda = e + fi$ where $a, b, c, d, e, f \in \mathbb{R}$.

**Left side — compute $(\alpha + \beta) + \lambda$:**

$$
(\alpha + \beta) + \lambda = \bigl((a + bi) + (c + di)\bigr) + (e + fi)
$$

$$
= \bigl((a+c) + (b+d)i\bigr) + (e + fi)
$$

$$
= \bigl((a+c) + e\bigr) + \bigl((b+d) + f\bigr)i
$$

**Right side — compute $\alpha + (\beta + \lambda)$:**

$$
\alpha + (\beta + \lambda) = (a + bi) + \bigl((c + di) + (e + fi)\bigr)
$$

$$
= (a + bi) + \bigl((c+e) + (d+f)i\bigr)
$$

$$
= \bigl(a + (c+e)\bigr) + \bigl(b + (d+f)\bigr)i
$$

Since real addition is associative, $(a+c)+e = a+(c+e)$ and $(b+d)+f = b+(d+f)$, so the two sides are equal.

Therefore $(\alpha + \beta) + \lambda = \alpha + (\beta + \lambda)$ for all $\alpha, \beta, \lambda \in \mathbb{C}$. $\blacksquare$

---

### Notes

I also noted in my scratchwork that in the left side calculation, you can bring in commutativity to rearrange into the right side form. The key step is that associativity and commutativity of reals together let you freely rearrange — the final result looks the same either way.

## Exercise 3 — Associativity of Multiplication

**Show that $(\alpha\beta)\lambda = \alpha(\beta\lambda)$ for all $\alpha, \beta, \lambda \in \mathbb{C}$.**

---

### My Approach

Same strategy. Write everything out, use the definition of complex multiplication, and let real arithmetic do the rest. The multiplication definition is messier than addition so there's more algebra, but the structure is identical.

One thing I flagged in my notes: when you expand $\alpha\beta$ you use $i^2 = -1$, which is built into the definition — $(a+bi)(c+di) = (ac - bd) + (ad+bc)i$.

### Proof

Let $\alpha = a + bi$, $\beta = c + di$, $\lambda = e + fi$ where $a, b, c, d, e, f \in \mathbb{R}$.

**First compute $\alpha\beta$:**

By the definition of complex multiplication:

$$
\alpha\beta = (a+bi)(c+di) = (ac - bd) + (ad + bc)i
$$

**Now compute $(\alpha\beta)\lambda$:**

$$
(\alpha\beta)\lambda = \bigl[(ac-bd) + (ad+bc)i\bigr](e + fi)
$$

$$
= \bigl[(ac-bd)e - (ad+bc)f\bigr] + \bigl[(ac-bd)f + (ad+bc)e\bigr]i
$$

Expanding:

$$
= (ace - bde - adf - bcf) + (acf - bdf + ade + bce)i
$$

**Now compute $\beta\lambda$:**

$$
\beta\lambda = (c+di)(e+fi) = (ce - df) + (cf + de)i
$$

**Now compute $\alpha(\beta\lambda)$:**

$$
\alpha(\beta\lambda) = (a+bi)\bigl[(ce-df) + (cf+de)i\bigr]
$$

$$
= \bigl[a(ce-df) - b(cf+de)\bigr] + \bigl[a(cf+de) + b(ce-df)\bigr]i
$$

Expanding:

$$
= (ace - adf - bcf - bde) + (acf + ade + bce - bdf)i
$$

The real parts match: $ace - bde - adf - bcf = ace - adf - bcf - bde$ ✓

The imaginary parts match: $acf - bdf + ade + bce = acf + ade + bce - bdf$ ✓

Therefore $(\alpha\beta)\lambda = \alpha(\beta\lambda)$ for all $\alpha, \beta, \lambda \in \mathbb{C}$. $\blacksquare$

---

### Notes

From my scratchwork I noted that invoking the distributive rule and using $i^2 = -1$ at each step is what drives the computation. Once everything is expanded into real terms you just check that both sides produce the same sum — commutativity of real multiplication handles rearranging the individual terms like $ace = ace$.

## Exercise 4 — Distributive Property

**Show that $\lambda(\alpha + \beta) = \lambda\alpha + \lambda\beta$ for all $\lambda, \alpha, \beta \in \mathbb{C}$.**

---

### Proof

Let $\lambda = e + fi$, $\alpha = a + bi$, $\beta = c + di$ where $a, b, c, d, e, f \in \mathbb{R}$.

**Left side:**

$$
\alpha + \beta = (a+c) + (b+d)i
$$

$$
\lambda(\alpha + \beta) = (e+fi)\bigl[(a+c) + (b+d)i\bigr]
$$

$$
= \bigl[e(a+c) - f(b+d)\bigr] + \bigl[f(a+c) + e(b+d)\bigr]i
$$

$$
= (ea + ec - fb - fd) + (fa + fc + eb + ed)i
$$

**Right side:**

$$
\lambda\alpha = (e+fi)(a+bi) = (ea - fb) + (fa + eb)i
$$

$$
\lambda\beta = (e+fi)(c+di) = (ec - fd) + (fc + ed)i
$$

$$
\lambda\alpha + \lambda\beta = (ea - fb + ec - fd) + (fa + eb + fc + ed)i
$$

Both sides give $(ea + ec - fb - fd) + (fa + fc + eb + ed)i$, so they're equal.

Therefore $\lambda(\alpha + \beta) = \lambda\alpha + \lambda\beta$ for all $\lambda, \alpha, \beta \in \mathbb{C}$. $\blacksquare$

## Exercise 5 — Additive Inverse (Unique)

**Show that for every $\alpha \in \mathbb{C}$, there exists a unique $\beta \in \mathbb{C}$ such that $\alpha + \beta = 0$.**

---

### My Approach

Two things to show: existence (give a $\beta$ that works) and uniqueness (show no other $\beta$ can work).

### Proof

Let $\alpha = a + bi$ where $a, b \in \mathbb{R}$.

**Existence:** Let $\beta = -a - bi$. Then:

$$
\alpha + \beta = (a + bi) + (-a - bi) = (a - a) + (b - b)i = 0 + 0i = 0
$$

So $\beta = -a - bi$ works.

**Uniqueness:** Suppose $\beta' \in \mathbb{C}$ also satisfies $\alpha + \beta' = 0$. Then:

$$
\beta' = \beta' + 0 = \beta' + (\alpha + \beta) = (\beta' + \alpha) + \beta = (\alpha + \beta') + \beta = 0 + \beta = \beta
$$

So $\beta' = \beta$, meaning the additive inverse is unique. $\blacksquare$

---

### Notes

The uniqueness argument is a standard trick — assume two things both satisfy the property, then chain equalities using those properties to show they must be the same object. I've seen this pattern in the proofs book too.

## Exercise 6 — Multiplicative Inverse (Unique)

**Show that for every $\alpha \in \mathbb{C}$ with $\alpha \neq 0$, there exists a unique $\beta \in \mathbb{C}$ such that $\alpha\beta = 1$.**

---

### Proof

Let $\alpha = a + bi$ where $a, b \in \mathbb{R}$ and $\alpha \neq 0$, so $(a, b) \neq (0, 0)$, meaning $a^2 + b^2 > 0$.

**Existence:** Let $\beta = \dfrac{a}{a^2 + b^2} - \dfrac{b}{a^2+b^2}i$. Then:

$$
\alpha\beta = (a+bi)\left(\frac{a}{a^2+b^2} - \frac{b}{a^2+b^2}i\right)
$$

$$
= \frac{1}{a^2+b^2}(a+bi)(a-bi) = \frac{a^2 + b^2}{a^2 + b^2} = 1
$$

**Uniqueness:** Suppose $\beta'$ also satisfies $\alpha\beta' = 1$. Then:

$$
\beta' = 1 \cdot \beta' = (\alpha\beta)\beta' = \beta(\alpha\beta') = \beta \cdot 1 = \beta
$$

So the multiplicative inverse is unique. $\blacksquare$

---

### Notes

The formula $\beta = \frac{\bar{\alpha}}{|\alpha|^2}$ is the standard complex reciprocal. The uniqueness argument is structurally the same as exercise 5 — assume two inverses exist, then show they're equal by multiplying through. Axler marks this one as easy to skip, and it kind of is once you've seen exercise 5.

## Exercise 7 — Cube Root of 1

**Show that $\dfrac{-1 + \sqrt{3}\,i}{2}$ is a cube root of 1.**

---

### Proof

Let $\omega = \dfrac{-1 + \sqrt{3}\,i}{2}$. We need to show $\omega^3 = 1$.

**Step 1 — Compute $\omega^2$:**

$$
\omega^2 = \left(\frac{-1 + \sqrt{3}\,i}{2}\right)^2 = \frac{(-1+\sqrt{3}\,i)^2}{4}
$$

$$
(-1+\sqrt{3}\,i)^2 = 1 - 2\sqrt{3}\,i + 3i^2 = 1 - 2\sqrt{3}\,i - 3 = -2 - 2\sqrt{3}\,i
$$

$$
\omega^2 = \frac{-2 - 2\sqrt{3}\,i}{4} = \frac{-1 - \sqrt{3}\,i}{2}
$$

**Step 2 — Compute $\omega^3 = \omega^2 \cdot \omega$:**

$$
\omega^3 = \frac{-1 - \sqrt{3}\,i}{2} \cdot \frac{-1 + \sqrt{3}\,i}{2} = \frac{(-1-\sqrt{3}\,i)(-1+\sqrt{3}\,i)}{4}
$$

The numerator is a difference of squares:

$$
(-1)^2 - (\sqrt{3}\,i)^2 = 1 - 3i^2 = 1 + 3 = 4
$$

$$
\omega^3 = \frac{4}{4} = 1
$$

Therefore $\omega$ is a cube root of 1. $\blacksquare$

## Exercise 8 — Two Distinct Square Roots of $i$

**Find two distinct square roots of $i$.**

---

### Approach

Want to find $z = a + bi$ such that $z^2 = i$.

$$
(a+bi)^2 = a^2 - b^2 + 2abi = 0 + 1 \cdot i
$$

Matching real and imaginary parts:

$$
a^2 - b^2 = 0 \quad \text{and} \quad 2ab = 1
$$

From the first equation: $a^2 = b^2$, so $a = b$ or $a = -b$.

**Case 1: $a = b$.** Then $2a^2 = 1$, so $a = \pm \dfrac{1}{\sqrt{2}}$.

**Case 2: $a = -b$.** Then $2ab = -2a^2 = 1$, so $a^2 = -\frac{1}{2}$, which has no real solution.

So the two square roots are:

$$
z = \frac{1}{\sqrt{2}} + \frac{1}{\sqrt{2}}i = \frac{\sqrt{2}}{2}(1 + i)
$$

$$
z = -\frac{1}{\sqrt{2}} - \frac{1}{\sqrt{2}}i = -\frac{\sqrt{2}}{2}(1 + i)
$$

## Exercise 10 — No Such $\lambda$ Exists

**Explain why there does not exist $\lambda \in \mathbb{C}$ such that**

$$
\lambda(2 - 3i,\ 5 + 4i,\ -6 + 7i) = (12 - 5i,\ 7 + 22i,\ -32 - 9i).
$$

---

### Solution

Scalar multiplication acts component-wise, so if such a $\lambda$ existed, it would have to satisfy all three equations simultaneously:

$$
\lambda(2 - 3i) = 12 - 5i
$$

$$
\lambda(5 + 4i) = 7 + 22i
$$

$$
\lambda(-6 + 7i) = -32 - 9i
$$

From the first equation:

$$
\lambda = \frac{12 - 5i}{2 - 3i} = \frac{(12 - 5i)(2 + 3i)}{(2-3i)(2+3i)} = \frac{24 + 36i - 10i - 15i^2}{4 + 9} = \frac{24 + 26i + 15}{13} = \frac{39 + 26i}{13} = 3 + 2i
$$

Check against the second equation:

$$
(3 + 2i)(5 + 4i) = 15 + 12i + 10i + 8i^2 = 15 + 22i - 8 = 7 + 22i \checkmark
$$

Check against the third equation:

$$
(3 + 2i)(-6 + 7i) = -18 + 21i - 12i + 14i^2 = -18 + 9i - 14 = -32 + 9i
$$

But we need $-32 - 9i$, and we got $-32 + 9i$. These are not equal.

So the $\lambda$ that satisfies the first two components fails the third, meaning no single $\lambda \in \mathbb{C}$ satisfies all three simultaneously. $\blacksquare$

## Exercise 11 — Associativity in $\mathbb{F}^n$

**Show that $(x + y) + z = x + (y + z)$ for all $x, y, z \in \mathbb{F}^n$.**

---

### Proof

Let $x = (x_1, \ldots, x_n)$, $y = (y_1, \ldots, y_n)$, $z = (z_1, \ldots, z_n)$ with $x_j, y_j, z_j \in \mathbb{F}$.

By the definition of addition in $\mathbb{F}^n$:

$$
((x+y)+z)_j = (x_j + y_j) + z_j
$$

Since addition in $\mathbb{F}$ is associative:

$$
(x_j + y_j) + z_j = x_j + (y_j + z_j) = (x + (y+z))_j
$$

Since this holds for every coordinate $j = 1, \ldots, n$, we have $(x+y)+z = x+(y+z)$. $\blacksquare$

## Exercise 12 — Scalar Multiplication Associativity in $\mathbb{F}^n$

**Show that $(ab)x = a(bx)$ for all $x \in \mathbb{F}^n$ and all $a, b \in \mathbb{F}$.**

---

### Proof

Let $x = (x_1, \ldots, x_n)$ with $x_j \in \mathbb{F}$.

By the definition of scalar multiplication:

$$
((ab)x)_j = (ab)x_j
$$

Since multiplication in $\mathbb{F}$ is associative:

$$
(ab)x_j = a(bx_j) = (a(bx))_j
$$

This holds for all $j$, so $(ab)x = a(bx)$. $\blacksquare$

## Exercise 13 — Multiplicative Identity in $\mathbb{F}^n$

**Show that $1x = x$ for all $x \in \mathbb{F}^n$.**

---

### Proof

Let $x = (x_1, \ldots, x_n)$. For each coordinate:

$$
(1x)_j = 1 \cdot x_j = x_j
$$

since $1$ is the multiplicative identity in $\mathbb{F}$. So $1x = x$. $\blacksquare$

## Exercise 14 — Distributivity over Vector Addition

**Show that $\lambda(x + y) = \lambda x + \lambda y$ for all $\lambda \in \mathbb{F}$ and all $x, y \in \mathbb{F}^n$.**

---

### Proof

Let $x = (x_1, \ldots, x_n)$, $y = (y_1, \ldots, y_n)$. For each coordinate:

$$
(\lambda(x+y))_j = \lambda(x_j + y_j) = \lambda x_j + \lambda y_j = (\lambda x + \lambda y)_j
$$

where the middle step uses distributivity in $\mathbb{F}$. Holds for all $j$, so $\lambda(x+y) = \lambda x + \lambda y$. $\blacksquare$

## Exercise 15 — Distributivity over Scalar Addition

**Show that $(a + b)x = ax + bx$ for all $a, b \in \mathbb{F}$ and all $x \in \mathbb{F}^n$.**

---

### Proof

Let $x = (x_1, \ldots, x_n)$. For each coordinate:

$$
((a+b)x)_j = (a+b)x_j = ax_j + bx_j = (ax + bx)_j
$$

where the middle step uses distributivity in $\mathbb{F}$. Holds for all $j$, so $(a+b)x = ax + bx$. $\blacksquare$
